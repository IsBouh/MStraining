```python
from nltk.tokenize import RegexpTokenizer  
from nltk.corpus import stopwords
```

```python
def process(input_text):
    
    liste_inter=[]
    liste_finale=[]
    
    
    for word in input_text:
        
        
        unaccented_string = unidecode.unidecode(word)
        #print(unaccented_string)

        
        # Create a regular expression tokenizer
        tokenizer = RegexpTokenizer(r'\w+')
        #print('token', tokenizer)

        # Create a Snowball stemmer 
        stemmer = SnowballStemmer('french')
        #print('stem', stemmer)

        # Get the list of stop words 
        stop_words = stopwords.words('french')
        #print('stop', stop_words)
    
        # Tokenize the input string
        tokens = tokenizer.tokenize(unaccented_string)

        # Remove the stop words 
        tokens = [x for x in tokens if not x in stop_words]
    
        # Perform stemming on the tokenized words 
        tokens_stemmed = [stemmer.stem(x) for x in tokens]
        #print('tok',tokens_stemmed)
        

        #print('lemma',tokens_stemmed)
        liste_inter.append(tokens_stemmed)
    
    #return liste_inter

    for elem in liste_inter:
        x=" ".join(elem)
        liste_finale.append(x)
    
    return liste_finale
        
```
